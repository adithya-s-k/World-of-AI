```python
#Importing the dependencies

import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection and Processing

```python
#Loading the dataset to Pandas DataFrame
loan_dataset = pd.read_csv('loan_prediction.csv')
```

```python
type(loan_dataset)
```

```python
#printing the first 5 rows of the data frame using head()
loan_dataset.head()
```

```python
# Number of rows and columns
loan_dataset.shape
```

```python
#statistical measure
loan_dataset.describe()
```

```python
#number of missing values in each column
loan_dataset.isnull().sum()
```

```python
# dropping the missing values
loan_dataset = loan_dataset.dropna()
```

```python
#number of missing values in each column
loan_dataset.isnull().sum()
```

```python
#label encoding (changes categorical value to numerical value)
loan_dataset.replace({"Loan_Status":{'N':0, 'Y':1}}, inplace=True)
```

```python
#printing the first 5 rows of the data frame using head()
loan_dataset.head()
```

```python
# Number of rows and columns
loan_dataset.shape
```

```python
#dependent column values
loan_dataset['Dependents'].value_counts()
```

```python
#replacing the value of 3+ to 4
loan_dataset = loan_dataset.replace(to_replace='3+', value=4)
```

```python
#dependent column values
loan_dataset['Dependents'].value_counts()
```

DATA VISUALIZATION

```python
#education & loan status
sns.countplot(x='Education', hue='Loan_Status', data=loan_dataset)
```

```python
#matial status & loan status
sns.countplot(x='Married', hue='Loan_Status', data=loan_dataset)
```

```python
#label encoding (changes categorical value to numerical value)
loan_dataset.replace({'Married':{'No':0, 'Yes':1}, 'Gender':{'Male':1, 'Female':0},
                     'Self_Employed':{'No':0,'Yes':1}, 'Property_Area':{'Rural':0, 'Semiurban':1,'Urban':2},
                     'Education':{'Graduate':1,'Not Graduate':0}}, inplace=True)
```

```python
#printing the first 5 rows of the data frame using head()
loan_dataset.head()
```

```python
#separating the data and label
X = loan_dataset.drop(columns = ['Loan_ID', 'Loan_Status'], axis = 1)
Y = loan_dataset['Loan_Status'] 
```

```python
print(X)
print(Y)
```

Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.1,stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Training the model

Support Vector Machine Model

```python
classifier = svm.SVC(kernel='linear')
```

```python
#training the support vector machine model
classifier.fit(X_train, Y_train)
```

Model Evaluation

```python
#accuracy score on training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on training data: ", training_data_accuracy)
```

```python
#accuracy score on test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on test data: ", test_data_accuracy)
```

Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train, Y_train)
y_predict_nb = nb.predict(X_test)
```

```python
y_predict_nb
```

```python
score = round(accuracy_score(y_predict_nb, Y_test)*100,2)
print("Accuracy using Naive Bayes: ", str(score)+"%")
```

K nearest neighbours

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train,Y_train)
Y_pred_knn=knn.predict(X_test)
```

```python
score_knn = round(accuracy_score(Y_pred_knn,Y_test)*100,2)

print("Accuracy on test data: "+str(score_knn)+" %")
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

max_accuracy = 0


for x in range(200):
    dt = DecisionTreeClassifier(random_state=x)
    dt.fit(X_train,Y_train)
    Y_pred_dt = dt.predict(X_test)
    current_accuracy = round(accuracy_score(Y_pred_dt,Y_test)*100,2)
    if(current_accuracy>max_accuracy):
        max_accuracy = current_accuracy
        best_x = x
        
#print(max_accuracy)
#print(best_x)


dt = DecisionTreeClassifier(random_state=best_x)
dt.fit(X_train,Y_train)
Y_pred_dt = dt.predict(X_test)
```

```python
score_dt = round(accuracy_score(Y_pred_dt,Y_test)*100,2)

print("Accuracy using decision tree: "+str(score_dt)+" %")
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

max_accuracy = 0


for x in range(2000):
    rf = RandomForestClassifier(random_state=x)
    rf.fit(X_train,Y_train)
    Y_pred_rf = rf.predict(X_test)
    current_accuracy = round(accuracy_score(Y_pred_rf,Y_test)*100,2)
    if(current_accuracy>max_accuracy):
        max_accuracy = current_accuracy
        best_x = x
        
#print(max_accuracy)
#print(best_x)

rf = RandomForestClassifier(random_state=best_x)
rf.fit(X_train,Y_train)
Y_pred_rf = rf.predict(X_test)
```

```python
score_rf = round(accuracy_score(Y_pred_rf,Y_test)*100,2)

print("Accuracy using random forest is : "+str(score_rf)+" %")
```
